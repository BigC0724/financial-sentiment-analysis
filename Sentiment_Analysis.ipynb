```python
# Import required libraries and load the sentiment analysis model
from transformers import pipeline

# Load FinBERT model for financial sentiment analysis
sentiment = pipeline("sentiment-analysis", 
                    model="ProsusAI/finbert", 
                    tokenizer="ProsusAI/finbert")

headlines_test = [
    "JPMorgan reports record quarterly profits",
    "Federal Reserve raises rates higher than expected",
    "Bank of America beats earnings expectations",
    "Credit card delinquencies rise to highest since 2008",
    "Goldman Sachs raises dividend by 25 percent",
    "OneMain Financial reports surge in loan defaults",
    "Consumer confidence reaches five year high",
    "Recession fears grow as consumer spending slows"
]

print("=" * 60)
print("FINBERT FINANCIAL SENTIMENT ANALYSIS")
print("=" * 60)

for headline in headlines_test:
    result = sentiment(headline)
    label = result[0]["label"]
    score = round(result[0]["score"] * 100, 1)
    
    # Add emoji based on sentiment
    if label == "positive":
        emoji = "📈"
    elif label == "negative":
        emoji = "📉"
    else:
        emoji = "➡️"
    
    print(f"\n{emoji} Headline: {headline}")
    print(f"   Sentiment: {label.upper()}")
    print(f"   Confidence: {score}%")

print("\n" + "=" * 60)
```

```python
# Run FinBERT on all 100 headlines
results = []

for headline in df["Headline"]:
    result = sentiment(headline)
    results.append({
        "Headline": headline,
        "FinBERT_Sentiment": result[0]["label"],
        "Confidence": round(result[0]["score"] * 100, 1)
    })

results_df = pd.DataFrame(results)

# Add original labels for comparison
results_df["Actual_Sentiment"] = df["Sentiment"].map(
    {1: "positive", 0: "negative"}
)

# Check accuracy
results_df["Correct"] = (
    results_df["FinBERT_Sentiment"] == 
    results_df["Actual_Sentiment"]
)

accuracy = results_df["Correct"].mean() * 100
print(f"📊 FinBERT Accuracy: {round(accuracy, 1)}%")
print(f"\n📈 Sentiment Distribution:")
print(results_df["FinBERT_Sentiment"].value_counts())
print(f"\n🔍 Sample Results:")
print(results_df[["Headline", "FinBERT_Sentiment", 
                   "Actual_Sentiment", 
                   "Confidence"]].head(10))
```

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# Chart 1 - FinBERT Sentiment Distribution
results_df["FinBERT_Sentiment"].value_counts().plot(
    kind="bar",
    ax=axes[0],
    color=["red", "blue", "gray"],
    alpha=0.7
)
axes[0].set_title("FinBERT Sentiment Distribution")
axes[0].set_xlabel("Sentiment")
axes[0].set_ylabel("Count")
axes[0].tick_params(axis="x", rotation=0)
axes[0].grid(True, alpha=0.3)

# Chart 2 - Accuracy by Sentiment Type
correct_by_sentiment = results_df.groupby(
    "Actual_Sentiment")["Correct"].mean() * 100

correct_by_sentiment.plot(
    kind="bar",
    ax=axes[1],
    color=["red", "blue"],
    alpha=0.7
)
axes[1].set_title("Accuracy by Sentiment Type")
axes[1].set_xlabel("Actual Sentiment")
axes[1].set_ylabel("Accuracy %")
axes[1].tick_params(axis="x", rotation=0)
axes[1].grid(True, alpha=0.3)

# Chart 3 - Confidence Distribution
results_df["Confidence"].hist(
    bins=20,
    ax=axes[2],
    color="purple",
    alpha=0.7
)
axes[2].set_title("FinBERT Confidence Distribution")
axes[2].set_xlabel("Confidence %")
axes[2].set_ylabel("Count")
axes[2].grid(True, alpha=0.3)

plt.suptitle("FinBERT Financial Sentiment Analysis Results",
             fontsize=14, fontweight="bold")
plt.tight_layout()
plt.savefig("finbert_results.png")
plt.show()
```
